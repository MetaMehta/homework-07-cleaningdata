## Oh, weird, ABV isn't a number. Convert it to a number for me, please.

It's going to take a few steps!

### First, let's just look at the ABV column by itself

```python
# 1. Convert the 'ABV' column to float
df['ABV'] = df['ABV'].str.replace('%', '').astype(float)

# 2. Count the unique ABV values and sort from highest to lowest
abv_counts = df['ABV'].value_counts().sort_values(ascending=False)
print(abv_counts)
```

### Hm, `%` isn't part of  a number. Let's remove it.

When you're confident you got it right, save the results back into the `ABV` column.

- *Tip: In programming the easiest way to remove something is to *replacing it with nothing*.*
- *Tip: "nothing" might seem like `NaN` sinc we talked about it a lot in class, but in this case it isn't! It's just an empty string, like ""*
- *Tip: `.replace` is used for replacing ENTIRE cells, while `.str.replace` is useful for replacing PARTS of cells (see my New York example)*

```python
df['ABV']
```

```python
# 1. Check the data type of the 'ABV' column
df['ABV'].dtypes

# 2. Convert the 'ABV' column to string
df['ABV'] = df['ABV'].astype(str)

# 3. Remove the '%' sign
df['ABV'] = df['ABV'].str.replace('%', '')

# 4. Convert the 'ABV' column back to numeric
df['ABV'] = df['ABV'].astype(float)

# 5. Check the updated 'ABV' column
print(df['ABV'])
```

### Now let's turn `ABV` into a numeric data type

Save the results back into the `ABV` column (again), and then check `df.dtypes` to make sure it worked.

```python
df['ABV'] = pd.to_numeric(df['ABV'], errors='coerce')
print(df.dtypes)
```

# Measuring International Bitterness Units

## Display all of the IPAs

Include American IPAs, Imperial IPAs, and anything else with "IPA in it."

IPA stands for [India Pale Ale](https://www.bonappetit.com/story/ipa-beer-styles), and is probably the most popular kind of beer in the US for people who are drinking [craft beer](https://www.craftbeer.com/beer/what-is-craft-beer).

```python
df[df['Style'].str.contains('IPA', na=False)]
```

IPAs are usually pretty hoppy and bitter. IBU stands for [International Bitterness Unit](http://www.thebrewenthusiast.com/ibus/), and while a lot of places like to brag about having the most bitter beer (it's an American thing!), IBUs don't necessary *mean anything*.

Let's look at how different beers have different IBU measurements.

## Draw the distribution of IBU measurements, but with *twenty* bins instead of the default of 10

- *Tip: Every time I ask for a distribution, I'm looking for a histogram*

```python

# Convert the 'IBUs' column to numeric, handling non-numeric values.
df['IBUs'] = pd.to_numeric(df['IBUs'], errors='coerce')

# Drop NaN values from the 'IBUs' column to avoid errors during plotting.
cleaned_IBUs = df['IBUs'].dropna()

# Plot the distribution of IBUs with 20 bins.
plt.figure(figsize=(10, 6))
plt.hist(cleaned_IBUs, bins=20, edgecolor='black')
plt.title('Distribution of IBU Measurements')
plt.xlabel('IBUs')
plt.ylabel('Frequency')
plt.show()
```

## Hm, Interesting distribution. List all of the beers with IBUs above the 75th percentile

- *Tip: There's a single command that gives you the 25/50/75th percentile*
- *Tip: You can just manually type the number when you list those beers*

```python

# Convert 'IBUs' to numeric, handling non-numeric values.
df['IBUs'] = pd.to_numeric(df['IBUs'], errors='coerce')

# Calculate the 75th percentile of the 'IBUs' column.
ibu_75th_percentile = df['IBUs'].quantile(0.75)

# Filter the DataFrame to show only beers with IBUs above the 75th percentile.
high_ibu_beers = df[df['IBUs'] > ibu_75th_percentile]

print(high_ibu_beers[['Beer', 'IBUs']])
```

```python
# 50th percentile of the 'IBUs' column.
ibu_50th_percentile = df['IBUs'].quantile(0.50)

# Filter the DataFrame to show only beers with IBUs above the 50th percentile.
high_ibu_beers = df[df['IBUs'] > ibu_50th_percentile]

print(high_ibu_beers[['Beer', 'IBUs']])
```

## List all of the beers with IBUs below the 25th percentile

```python
# Calculate the 25th percentile of the 'IBUs' column.
ibu_25th_percentile = df['IBUs'].quantile(0.25)

# Filter the DataFrame to show only beers with IBUs above the 25th percentile.
high_ibu_beers = df[df['IBUs'] > ibu_25th_percentile]

# Assuming 'Beer' is the column with the beer names.
print(high_ibu_beers[['Beer', 'IBUs']])
```

## List the median IBUs of each type of beer. Graph it.

Put the highest at the top, and the missing ones at the bottom.

- Tip: Look at the options for `sort_values` to figure out the `NaN` thing. The `?` probably won't help you here.

```python

# Convert 'IBUs' to numeric, handling non-numeric values.
df['IBUs'] = pd.to_numeric(df['IBUs'], errors='coerce')

# Drop the null values from the 'IBUs' column.
non_null_ibus = df['IBUs'].dropna()

# Group by 'Style' and calculate the median IBUs for non-null values.
median_ibus = non_null_ibus.groupby(df['Style']).median()

# Sort the values, with the highest at the top.
median_ibus_sorted = median_ibus.sort_values(ascending=False)

# Plotting the results with a larger figure size to spread out the x-axis items.
plt.figure(figsize=(14, 10))
median_ibus_sorted.plot(kind='bar', linewidth=3, width=0.8)  # Adjust bar thickness and spacing
plt.title('Median IBUs of Each Beer Style (Excluding Nulls)')
plt.xlabel('Beer Style')
plt.ylabel('Median IBUs')
plt.xticks(rotation=90)

# Adjust the margins to increase space between x-axis names.
plt.margins(x=0.05)  # Adjust the x-axis margins.

plt.tight_layout()  # Adjust the layout to fit the plot and labels.
plt.show()


```

## Hmmmm, it looks like they are generally different styles. What are the most common 5 styles of high-IBU beer vs. low-IBU beer?

- *Tip: You'll want to think about it in three pieces - filtering to only find the specific beers beers, then finding out what the most common styles are, then getting the top 5.*
- *Tip: You CANNOT do this in one command. It's going to be one command for the high and one for the low.*
- *Tip: "High IBU" means higher than 75th percentile, "Low IBU" is under 25th percentile*

```python

# Convert 'IBUs' to numeric, handling non-numeric values.
df['IBUs'] = pd.to_numeric(df['IBUs'], errors='coerce')

# The 25th and 75th percentiles of the 'IBUs' column.
ibu_25th_percentile = df['IBUs'].quantile(0.25)
ibu_75th_percentile = df['IBUs'].quantile(0.75)

# Filter the DataFrame for high IBU beers (above the 75th percentile).
high_ibu_beers = df[df['IBUs'] > ibu_75th_percentile]

# Five the most common styles among high IBU beers.
top_5_high_ibu_styles = high_ibu_beers['Style'].value_counts().head(5)

# Filter the DataFrame for low IBU beers (below the 25th percentile).
low_ibu_beers = df[df['IBUs'] < ibu_25th_percentile]

# Five most common styles among low IBU beers.
top_5_low_ibu_styles = low_ibu_beers['Style'].value_counts().head(5)

# Display the results.
print("Top 5 High IBU Beer Styles:\n", top_5_high_ibu_styles)
print("\nTop 5 Low IBU Beer Styles:\n", top_5_low_ibu_styles)
```

## Get the average IBU of "Witbier", "Hefeweizen" and "American Pale Wheat Ale" styles

I'm counting these as wheat beers. If you see any other wheat beer categories, feel free to include them. I want ONE measurement and ONE graph, not three separate ones. And 20 to 30 bins in the histogram, please.

- *Tip: I hope that `isin` is in your toolbox*

```python

# DataFrame for the specified wheat beer styles.
wheat_beers_styles = ['Witbier', 'Hefeweizen', 'American Pale Wheat Ale']
wheat_beers = df[df['Style'].isin(wheat_beers_styles)]

# Convert 'IBUs' to numeric, handling non-numeric values.
wheat_beers['IBUs'] = pd.to_numeric(wheat_beers['IBUs'], errors='coerce')

# Average IBU for the wheat beer styles.
average_ibu = wheat_beers['IBUs'].mean()
print(f"The average IBU for the wheat beer styles is: {average_ibu:.2f}")

# Histogram with 20 to 30 bins.
plt.figure(figsize=(10, 6))
plt.hist(wheat_beers['IBUs'].dropna(), bins=25, edgecolor='black')  # You can adjust the number of bins here
plt.title('Distribution of IBUs for Wheat Beer Styles')
plt.xlabel('IBUs')
plt.ylabel('Frequency')
plt.show()
```

## Draw a histogram of the IBUs of those beers

```python

# Filter the DataFrame for the specified wheat beer styles.
wheat_beers_styles = ['Witbier', 'Hefeweizen', 'American Pale Wheat Ale']
wheat_beers = df[df['Style'].isin(wheat_beers_styles)].copy()

# Convert 'IBUs' to numeric, handling non-numeric values using .loc to avoid SettingWithCopyWarning.
wheat_beers.loc[:, 'IBUs'] = pd.to_numeric(wheat_beers['IBUs'], errors='coerce')

# Drop NaN values from the 'IBUs' column to avoid errors during plotting.
cleaned_IBUs = wheat_beers['IBUs'].dropna()

# Histogram with 20 to 30 bins.
plt.figure(figsize=(10, 6))
plt.hist(cleaned_IBUs, bins=25, edgecolor='black')  # Adjust the number of bins as needed
plt.title('Histogram of IBUs for Wheat Beer Styles')
plt.xlabel('IBUs')
plt.ylabel('Frequency')
plt.show()
```

## Get the average IBU of any style with "IPA" in it (also draw a histogram)

```python

# Filter the DataFrame for styles containing "IPA".
ipa_beers = df[df['Style'].str.contains("IPA", na=False)].copy()

# Convert 'IBUs' to numeric, handling non-numeric values using .loc to avoid SettingWithCopyWarning.
ipa_beers.loc[:, 'IBUs'] = pd.to_numeric(ipa_beers['IBUs'], errors='coerce')

# Average IBU for IPA styles.
average_ibu_ipa = ipa_beers['IBUs'].mean()
print(f"The average IBU for IPA styles is: {average_ibu_ipa:.2f}")

# Histogram with 20 to 30 bins for the IBU values of IPA styles.
plt.figure(figsize=(10, 6))
plt.hist(ipa_beers['IBUs'].dropna(), bins=25, edgecolor='black')  # Adjust the number of bins as needed
plt.title('Histogram of IBUs for IPA Beer Styles')
plt.xlabel('IBUs')
plt.ylabel('Frequency')
plt.show()
```

## Plot those two histograms in the same cell!

I swear the behavior of this changes every year. Does it make two charts? Does it overlap them in one?

```python

# Filter for wheat beers and IPAs, and convert 'IBUs' to numeric.
wheat_beers_styles = ['Witbier', 'Hefeweizen', 'American Pale Wheat Ale']
wheat_beers = df[df['Style'].isin(wheat_beers_styles)].copy()
ipa_beers = df[df['Style'].str.contains("IPA", na=False)].copy()

wheat_beers.loc[:, 'IBUs'] = pd.to_numeric(wheat_beers['IBUs'], errors='coerce')
ipa_beers.loc[:, 'IBUs'] = pd.to_numeric(ipa_beers['IBUs'], errors='coerce')

# Drop NaN values from the 'IBUs' column to avoid errors during plotting.
cleaned_IBUs_wheat = wheat_beers['IBUs'].dropna()
cleaned_IBUs_ipa = ipa_beers['IBUs'].dropna()

# Average IBU for wheat beers and IPAs.
average_ibu_wheat = cleaned_IBUs_wheat.mean()
average_ibu_ipa = cleaned_IBUs_ipa.mean()
print(f"The average IBU for wheat beer styles is: {average_ibu_wheat:.2f}")
print(f"The average IBU for IPA styles is: {average_ibu_ipa:.2f}")

# Histogram with 20 to 30 bins for both wheat beers and IPAs.
plt.figure(figsize=(10, 6))
plt.hist(cleaned_IBUs_wheat, bins=25, edgecolor='black', alpha=0.5, label='Wheat Beers')
plt.hist(cleaned_IBUs_ipa, bins=25, edgecolor='black', alpha=0.5, label='IPAs')
plt.title('Histogram of IBUs for Wheat Beer Styles and IPAs')
plt.xlabel('IBUs')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Compare the ABV of wheat beers vs. IPAs : their IBUs were really different, but how about their alcohol percentage?

Wheat beers might include witbier, hefeweizen, American Pale Wheat Ale, and anything else you think is wheaty. IPAs probably have "IPA" in their name.

```python

# Turn'ABV' column to string type.
df['ABV'] = df['ABV'].astype(str)

# Filter for wheat beers and IPAs.
wheat_beers_styles = ['Witbier', 'Hefeweizen', 'American Pale Wheat Ale']
wheat_beers = df[df['Style'].isin(wheat_beers_styles)]
ipa_beers = df[df['Style'].str.contains("IPA", na=False)]

# Remove the '%' sign and convert 'ABV' to numeric, handling non-numeric values.
wheat_beers['ABV'] = pd.to_numeric(wheat_beers['ABV'].str.replace('%', ''), errors='coerce')
ipa_beers['ABV'] = pd.to_numeric(ipa_beers['ABV'].str.replace('%', ''), errors='coerce')

# Calculate the average ABV for wheat beers and IPAs.
average_abv_wheat = wheat_beers['ABV'].mean()
average_abv_ipa = ipa_beers['ABV'].mean()
print(f"The average ABV for wheat beer styles is: {average_abv_wheat:.2f}%")
print(f"The average ABV for IPA styles is: {average_abv_ipa:.2f}%")

# Create histograms for the ABV values of wheat beers and IPAs.
plt.figure(figsize=(10, 6))
plt.hist(wheat_beers['ABV'].dropna(), bins=25, edgecolor='black', alpha=0.5, label='Wheat Beers')
plt.hist(ipa_beers['ABV'].dropna(), bins=25, edgecolor='black', alpha=0.5, label='IPAs')
plt.title('Histogram of ABV for Wheat Beer Styles and IPAs')
plt.xlabel('ABV (%)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
